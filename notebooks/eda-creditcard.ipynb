```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

# Load dataset

```python
credit = pd.read_csv("../data/raw/creditcard.csv")
```

# Dataset overview

```python
credit.head()
credit.info()
credit.describe()
```

# Class imbalance

```python
credit['Class'].value_counts()
```

```python
credit['Class'].value_counts(normalize=True) * 100
```

```python
sns.countplot(x='Class', data=credit)
plt.title("Credit Card Fraud Class Distribution")
plt.show()
```

# Amount vs Fraud

```python
sns.boxplot(x='Class', y='Amount', data=credit)
plt.title("Transaction Amount vs Fraud")
plt.show()
```

# Time vs Fraud

```python
sns.histplot(data=credit, x='Time', hue='Class', bins=50)
plt.title("Transaction Time Distribution by Class")
plt.show()
```
